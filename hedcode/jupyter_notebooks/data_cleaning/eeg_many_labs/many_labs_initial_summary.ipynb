## Initial summary of event files

```python
from hed.tools import  get_file_list, TabularSummary

# Variables to set for the specific dataset
#bids_root_path = '/openneuro/ds002790-download'
bids_root_path = 'O:\BIDSManyLabs\BIDS_Raw'
exclude_dirs = ['derivatives', 'models']
skip_columns = ['onset', 'sample']
# tasks = ['stopsignal']
name = 'eeg_many_labs_pilot'
# Construct the event file dictionaries for the BIDS and for EEG.event files
event_files = get_file_list(bids_root_path, extensions=[".tsv"], name_suffix="_events",
                            exclude_dirs=exclude_dirs)
file_dict = {file_path: file_path for file_path in event_files}
```

```python
print('\nBIDS events summary counts:')
summary_all = TabularSummary(skip_cols=skip_columns)
for path in event_files:
    this_summary = TabularSummary(skip_cols=skip_columns)
    this_summary.update(path)
    print(f"\nBIDS-style event info {path}:\n{this_summary}")
    summary_all.update(path)
print(f"\nBIDS-style event info for all:\n{summary_all}")
```
